# 3.Anatomy of Matplotlib Figure  <a class="anchor" id="3"></a>

[Table of Contents](#0.1)

This figure shows the name of several matplotlib elements composing a figure

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter


np.random.seed(19680801)

X = np.linspace(0.5, 3.5, 100)
Y1 = 3+np.cos(X)
Y2 = 1+np.cos(1+X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)


def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x

ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which='major', width=1.0)
ax.tick_params(which='major', length=10)
ax.tick_params(which='minor', width=1.0, labelsize=10)
ax.tick_params(which='minor', length=5, labelsize=10, labelcolor='0.25')

ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

ax.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(X, Y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='k')

ax.set_title("Anatomy of a figure", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")

ax.legend()


def circle(x, y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x, y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax.add_artist(circle)


def text(x, y, text):
    ax.text(x, y, text, backgroundcolor="white",
            ha='center', va='top', weight='bold', color='blue')


# Minor tick
circle(0.50, -0.10)
text(0.50, -0.32, "Minor tick label")

# Major tick
circle(-0.03, 4.00)
text(0.03, 3.80, "Major tick")

# Minor tick
circle(0.00, 3.50)
text(0.00, 3.30, "Minor tick")

# Major tick label
circle(-0.15, 3.00)
text(-0.15, 2.80, "Major tick label")

# X Label
circle(1.80, -0.27)
text(1.80, -0.45, "X axis label")

# Y Label
circle(-0.27, 1.80)
text(-0.27, 1.6, "Y axis label")

# Title
circle(1.60, 4.13)
text(1.60, 3.93, "Title")

# Blue plot
circle(1.75, 2.80)
text(1.75, 2.60, "Line\n(line plot)")

# Red plot
circle(1.20, 0.60)
text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(3.20, 1.75)
text(3.20, 1.55, "Markers\n(scatter plot)")

# Grid
circle(3.00, 3.00)
text(3.00, 2.80, "Grid")

# Legend
circle(3.70, 3.80)
text(3.70, 3.60, "Legend")

# Axes
circle(0.5, 0.5)
text(0.5, 0.3, "Axes")

# Figure
circle(-0.3, 0.65)
text(-0.3, 0.45, "Figure")

color = 'blue'
ax.annotate('Spines', xy=(4.0, 0.35), xycoords='data',
            xytext=(3.3, 0.5), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.annotate('', xy=(3.15, 0.0), xycoords='data',
            xytext=(3.45, 0.45), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.text(4.0, -0.4, "Made with http://matplotlib.org",
        fontsize=10, ha="right", color='.5')

plt.show()
````

The figure contains the overall window where plotting happens, contained within the figure are where actual graphs are plotted. Every Axes has an x-axis and y-axis for plotting. And contained within the axes are titles, ticks, labels associated with each axis. An essential figure of matplotlib is that we can more than axes in a figure which helps in building multiple plots, as shown below. In matplotlib, pyplot is used to create figures and change the characteristics of figures.

Bar Graph is used when to compare between between two categories or groups.

````python
## Basic Plot

plt.figure(figsize=(10,5))
x_ax = np.arange(1,10)
y_ax = np.arange(2000,11000,1000)
plt.bar(x_ax,y_ax)
plt.xlabel('Customer ID')
plt.ylabel('Sales Value')
plt.show()
````

### 5.1 Setting Color <a class="anchor" id="5.1"></a>

[Table of Contents](#0.1)


````python
plt.figure(figsize=(10,5)) # Setting the figure size
x_ax = np.arange(1,10)
y_ax = np.arange(2000,11000,1000)
ax = plt.axes() 
plt.bar(x_ax,y_ax,color = '#FFA726')
plt.xlabel('Customer ID')
plt.ylabel('Sales Value')
plt.show()
````

### 5.2 Add Label <a class="anchor" id="5.2"></a>

[Table of Contents](#0.1)


````python
x_ax = np.arange(1,10)
y_ax = np.arange(2000,11000,1000)
plt.figure(figsize=(10,5))
barplt = plt.bar(x_ax, y_ax)

# Make labels
for rect in barplt:
    height = rect.get_height()
    plt.text(
        rect.get_x() + rect.get_width() / 2, height + 5, int(height), ha="center", va="bottom"
    )
    
plt.title("Title")
plt.xlabel('Customer ID')
plt.ylabel("Sales Value")
plt.show()
````

### 5.3 Multiple Categories & Group Bar Chart <a class="anchor" id="5.3"></a>

[Table of Contents](#0.1)


````python
x1= [1,3,5,7]
x2=[2,4,6,8]
y1 = [10,10,10,10]
y2= [10,10,10,10]
plt.figure(figsize=(10,6))
ax = plt.axes()
ax.set_facecolor("white")
plt.bar(x1,y1,label = "First",color = '#3b7bf4') # First set of data
plt.bar(x2,y2,label = "Second",color = '#FFA726') # Second set of data
plt.xlabel('X')
plt.ylabel('Y')
plt.title ('Multiple Categories')
plt.legend()
plt.show()
````

````python
# Grouped Bar Chart
plt.figure(figsize=(7,9))

# set width of bar
barWidth = 0.25

x_ax = np.arange(1,10)

# set height of bar
y_ax1= np.array([18,27,30,35])
y_ax2 =np.array([22,31,21,19])
y_ax3 =np.array([10,21,31,21])
# Set position of bar on X axis
pos1 = np.arange(len(y1))
pos2 = [x + barWidth for x in pos1]
pos3 = [x + barWidth for x in pos2]
# Make the plot
plt.bar(pos1, y_ax1, color='#FF6F00', width=barWidth, label='Bar 1')
plt.bar(pos2, y_ax2, color='#FFA726', width=barWidth, label='Bar 2')
plt.bar(pos3, y_ax3, color='#D00000', width=barWidth, label='Bar 3')
# Add xticks on the middle of the group bars
plt.xlabel('Name', fontweight='bold')
plt.ylabel('Number', fontweight='bold')
plt.xticks([i + barWidth for i in range(len(y1))], ['User 1', 'User 2', 'User 3', 'User 4'])


for x,y in zip(pos1,y_ax1):
    plt.text(x, y, '%d' % y, ha='center' , va= 'bottom')
for x,y in zip(pos2,y_ax2):
    plt.text(x, y, '%d' % y, ha='center' , va= 'bottom')
for x,y in zip(pos3,y_ax3):
    plt.text(x, y, '%d' % y, ha='center' , va= 'bottom')
plt.title ('$Grouped $ $ Bar $ $ Chart$')

# Create legend & Show graphic
plt.legend()
plt.show()
````

### 5.4 Horizontal Bar <a class="anchor" id="5.4"></a>

[Table of Contents](#0.1)


#### Normal Horizontal Bar

````python
x_ax = np.arange(1,10)
y_ax = np.arange(2000,11000,1000)

plt.figure(figsize = (10,8))
plt.barh(x_ax,y_ax)
plt.title("Horizontal Bar")
plt.ylabel('Customer ID')
plt.xlabel("Sales Value")
plt.show()
````

#### Reverse Horizontal Bar

````python
x_ax = np.arange(1,10)
y_ax = np.arange(2000,11000,1000)

plt.figure(figsize = (10,8))
plt.barh(x_ax,y_ax)
plt.title("Reverse Horizontal Bar")
plt.ylabel('Customer ID')
plt.xlabel("Sales Value")
ax=plt.subplot()
ax.invert_yaxis()
plt.show()
````

````python
x_ax= ['User 1','User 2','User 3','User 4']
y_ax1= np.array([18,27,30,35])
y_ax2 =np.array([22,31,21,30])
plt.figure(figsize=(8,5))
plt.barh(x_ax,y_ax1,label = "Bar 1",color = '#FF6F00')
plt.barh(x_ax,y_ax2,label = "Bar 2", left = y_ax1 , color = '#FFB300')
plt.xlabel('Name')
plt.ylabel('Number')
plt.title ('Horizontal Bar Chart')
for x,y in zip(x_ax,y_ax1):
    plt.text(y-10, x, '%d' % y, ha='center' , va= 'bottom')
for x,y,z in zip(x_ax,y_ax2,y_ax1):
    plt.text(y+z-10, x, '%d' % y, ha='center' , va= 'bottom')
plt.title ('$Bar $ $ Chart$')
plt.legend()
plt.show()
````

### 5.5 Stacked Bar Chart <a class="anchor" id="5.5"></a>

[Table of Contents](#0.1)


````python
x_ax = np.arange(1,21)
plt.figure(figsize=(16,8))
y_ax1 = np.random.uniform(0.1,0.7,20)
y_ax2 = np.random.uniform(0.1,0.7,20)
plt.bar(x_ax, +y_ax1, facecolor='#3b7bf4', edgecolor='white') #specify edgecolor by name
plt.bar(x_ax, -y_ax2, facecolor='#FFA726', edgecolor='white')
for x,y in zip(x_ax,y_ax1):
    plt.text(x, y+0.05, '%.2f' % y, ha='center' , va= 'bottom', fontsize = 10)
    plt.xlim(0,21)
    plt.ylim(-1.25,+1.25)
plt.show()
````

````python
x_ax= ['User 1','User 2','User 3','User 4']
y_ax1 = [180,270,300,350]
y_ax2 = [20,30,20,20]
plt.figure(figsize=(10,8))
plt.bar(x_ax,y_ax1,label = "Sales Value",width = 0.5,color = '#FF6F00')
plt.bar(x_ax,y_ax2,label = "Quantity",width = 0.5 ,bottom = y_ax1 , color = '#FFA726')
plt.xlabel('Customer ID')
plt.ylabel('Sales Value x Quantity')
plt.title ('Stacked Bar')
for x,y in zip(x_ax,y_ax1):
    plt.text(x, y-100, '%d' % y, ha='center' , va= 'bottom')
for x,y,z in zip(x_ax,y_ax2,y_ax1):
    plt.text(x, y+z-10, '%d' % y, ha='center' , va= 'bottom')
plt.legend()
plt.show()
````
